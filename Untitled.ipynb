```python

df=pd.read_csv('perrin-freres-monthly-champagne- copy.csv')
df.tail()
df.drop(106,axis=0,inplace=True)
df.drop(105,axis=0,inplace=True)

df.columns=['Month','sales_per_month']
df.tail()
```

```python
df['Month']=pd.to_datetime(df['Month'])
df.head(14)
```

```python
df.set_index('Month',inplace=True)
df.head(14)
df.tail()
```

```python
df.plot()
```

```python
model=sm.tsa.statespace.SARIMAX(df['sales_per_month'],order=(1,0,0),seasonal_order=(1,1,1,12))
results=model.fit()
```

```python
df['forecast']=results.predict(start=90,end=104,dynamic=True)
print(df['forecast'].tail())
df[['sales_per_month','forecast']].plot(figsize=(12,8))
```
